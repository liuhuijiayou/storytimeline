```python
import pymysql
import re
import json
conn = pymysql.connect(port=3306,user='root',passwd='220155',charset='utf8',database='weibo_media')
cursor = conn.cursor()
sql = 'select post_content,post_time from post order by post_time'
cursor.execute(sql)
weibo_media_data = cursor.fetchall()
```

```python
tag_list = {}
no_tag = []
```

```python
for i,doc in enumerate(weibo_media_data):
    if i==900:
        break
    sentence = doc[0]
    # 提取话题
    # topics = re.findall("<#>.*?</#>",data)
    topics = re.findall("(?<=<#>).*?(?=</#>)", sentence)
    if len(topics)==0:
        no_tag.append(i)
        continue;
    for topic in topics:
        if topic not in tag_list:
            tag_list[topic] = [i]
        else:
            tag_list[topic].append(i)
```

```python
with open('reference.json','w',encoding='utf-8') as f:
    json.dump(tag_list,f,ensure_ascii=False)
    json.dump(no_tag,f,ensure_ascii=False)
```
