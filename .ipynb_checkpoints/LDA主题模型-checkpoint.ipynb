```python
def process(process_data):
    for c, sentence in enumerate(process_data):
        data = sentence[0]
        # print(data)
        # 提取话题
        # topics = re.findall("<#>.*?</#>",data)
        topics = re.findall("(?<=<#>).*?(?=</#>)", data)
        data = re.sub("<#>.*?</#>", '', data)
        # print('话题:',topics)
        # 提取链接
        # links = re.findall("<u>.*?</u>", data)
        links = re.findall("(?<=<u>).*?(?=</u>)", data)
        data = re.sub("<u>.*?</u>", '', data)
        # print('链接:', links)
        # 提取用户
        # users = re.findall("<@>.*?</@>", data)
        users = re.findall("(?<=<@>).*?(?=</@>)", data)
        data = re.sub("<@>.*?</@>", '', data)
        # print('用户:', users)
        # print(data)
        # 数据清洗
        # 去除英文
        data = re.sub("[^\u4e00-\u9fa5]", '', data)
        # 去除无用词
        #data = re.sub("表示|说|分享|转发|快转|微博|快转微博|转发微博|分享图片|分享", '', data)
        # 分词识别词性

        words = CRFnewSegment.seg(data)
        word_frequency = {}
        for term in words:
            word = str(term.word)
            nature = str(term.nature)
            if nature in Part_of_speechs or word in stop_words:
                continue
            # 统计当前帖子的词频
            if word in word_frequency:
                word_frequency[word] = word_frequency[word] + 1
            else:
                word_frequency[word] = 1

            # 统计总词频
            if word in wordcount:
                wordcount[word] = wordcount[word] + 1
            else:
                wordcount[word] = 1
                '''#构建词典
                if word not in word2num:
                    word2num[word] = count
                    num2word[count] = word
                    count = count + 1'''
        # topic处理
        topic_frequency = {}
        for i in topics:
            i = str(i)
            if i in topic_frequency:
                topic_frequency[i] = topic_frequency[i] + 1
            else:
                topic_frequency[i] = 1
            if i in topiccount:
                topiccount[i] = topiccount[i] + 1
            else:
                topiccount[i] = 1

        link_frequency = {}
        # link处理
        for i in links:
            i = str(i)
            value = re.findall('.*?(?=:\[\'.*?\'\])', i)[0]
            value = re.sub("[^\u4e00-\u9fa5]", '', value)
            i = re.findall('(?<=\[\').*?(?=\'\])', i)[0]  # url
            value_url = str(value) + '_' + str(i)
            if i in link_frequency:
                link_frequency[i] = link_frequency[i] + 1
            else:
                link_frequency[i] = 1
            if i in linkcount:
                linkcount[i] = linkcount[i] + 1
            else:
                linkcount[i] = 1
            if value_url in value_url_count:
                value_url_count[value_url] = value_url_count[value_url] + 1
            else:
                value_url_count[value_url] = 1

        # 计算词与词共现频率
        for i1, v1 in word_frequency.items():
            for i2, v2 in word_frequency.items():
                if i1 != i2:
                    co_entities = i1 + '(/-)0(/-)' + i2 + '(/-)0'
                    co_frequency = v1 * v2
                    if co_entities in co_ccurrence_matrix:
                        co_ccurrence_matrix[co_entities] = co_ccurrence_matrix[co_entities] + co_frequency
                    else:
                        co_ccurrence_matrix[co_entities] = co_frequency
        # 计算topic与topic共现频率
        for i1, v1 in topic_frequency.items():
            for i2, v2 in topic_frequency.items():
                if i1 != i2:
                    co_entities = i1 + '(/-)1(/-)' + i2 + '(/-)1'
                    co_frequency = v1 * v2
                    if co_entities in co_ccurrence_matrix:
                        co_ccurrence_matrix[co_entities] = co_ccurrence_matrix[co_entities] + co_frequency
                    else:
                        co_ccurrence_matrix[co_entities] = co_frequency

        # 计算link与link共现频率
        for i1, v1 in link_frequency.items():
            for i2, v2 in link_frequency.items():
                if i1 != i2:
                    co_entities = i1 + '(/-)2(/-)' + i2 + '(/-)2'
                    co_frequency = v1 * v2
                    if co_entities in co_ccurrence_matrix:
                        co_ccurrence_matrix[co_entities] = co_ccurrence_matrix[co_entities] + co_frequency
                    else:
                        co_ccurrence_matrix[co_entities] = co_frequency

        # 计算word与topic共现频率
        for i1, v1 in word_frequency.items():
            for i2, v2 in topic_frequency.items():
                co_entities = i1 + '(/-)0(/-)' + i2 + '(/-)1'
                co_frequency = v1 * v2
                if co_entities in co_ccurrence_matrix:
                    co_ccurrence_matrix[co_entities] = co_ccurrence_matrix[co_entities] + co_frequency
                else:
                    co_ccurrence_matrix[co_entities] = co_frequency

        # 计算word与link共现频率
        for i1, v1 in word_frequency.items():
            for i2, v2 in link_frequency.items():
                co_entities = i1 + '(/-)0(/-)' + i2 + '(/-)2'
                co_frequency = v1 * v2
                if co_entities in co_ccurrence_matrix:
                    co_ccurrence_matrix[co_entities] = co_ccurrence_matrix[co_entities] + co_frequency
                else:
                    co_ccurrence_matrix[co_entities] = co_frequency
        # 计算topic与link共现频率
        for i1, v1 in topic_frequency.items():
            for i2, v2 in link_frequency.items():
                co_entities = i1 + '(/-)1(/-)' + i2 + '(/-)2'
                co_frequency = v1 * v2
                if co_entities in co_ccurrence_matrix:
                    co_ccurrence_matrix[co_entities] = co_ccurrence_matrix[co_entities] + co_frequency
                else:
                    co_ccurrence_matrix[co_entities] = co_frequency
        if c % 1000 == 0:
            print(c)
process(weibo_data)
```

```python
from gensim.models import LdaModel
from gensim.corpora import Dictionary
doc_clean = [doc.split() for doc in doc_complete]
```

```python
# 创建语料的词语词典，每个单独的词语都会被赋予一个索引
dictionary = Dictionary(doc_clean)#id2word

# 使用上面的词典，将转换文档列表（语料）变成 DT 矩阵，文档转换成词袋模型
doc_term_matrix = [dictionary.doc2bow(doc) for doc in doc_clean]
```

```python
# 在 DT 矩阵上运行和训练 LDA 模型
ldamodel = LdaModel(doc_term_matrix, num_topics=3, id2word = dictionary, passes=50)
```

```python
# 输出结果
print(ldamodel.print_topics(num_topics=3, num_words=3))
```
