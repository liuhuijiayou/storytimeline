```python
sql = 'select post_content from post order by post_time'
cursor.execute(sql)
weibo_media_data = cursor.fetchall()
```

```python
NS = set()#地点名
NR = set()#人名
NT = set()#机构名
for data in weibo_media_data:
    data = data[-1]
    topics = re.findall("(?<=<#>).*?(?=</#>)", data)
    links = re.findall("(?<=<u>).*?(?=</u>)", data)
    users = re.findall("(?<=<@>).*?(?=</@>)", data)
    title = re.findall("(?<=【).*?(?=】)",data)
    data = re.sub("<#>.*?</#>|<u>.*?</u>|<@>.*?</@>|【.*?】", '', data)
    data = re.sub('据.{1,15}报道|()','',data)
    sentences = re.split('(?:[。|！|\!||？|\?|；|;])',data)
    for sentence in sentences:
        #print(sentence)
        ne_s = recognizer(sentence)
        for ne in ne_s:
            if ne[1]=='NS':
                NS.add(ne[0])
            if ne[1]=='NT':
                NT.add(ne[0])
            if ne[1]=='NR':
                NR.add(ne[0])
        #print(ne_s)
    print('-'*100)
    #print(data)
```
